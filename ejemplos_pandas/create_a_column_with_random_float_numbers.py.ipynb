```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame()

N = 1000
df['rand_col_1'] = np.random.RandomState(0).uniform(0, 1, size=N)
df['rand_col_2'] = np.random.RandomState(0).normal(0, 1, size=N)
df['rand_col_3'] = np.random.RandomState(0).binomial(10, 0.5, size=N)
df['rand_col_4'] = np.random.RandomState(1).poisson(3, size=N)

df.hist(bins=10)
plt.show()

#Este código crea un DataFrame de Pandas con cuatro columnas (rand_col_1, rand_col_2, rand_col_3, rand_col_4) 
#que contienen números aleatorios generados con diferentes distribuciones utilizando la librería NumPy. 
#Luego, se crea un histograma de cada una de estas columnas con 10 bins
#y se muestra en una ventana gráfica utilizando matplotlib.pyplot.
#En resumen, este código muestra visualmente la distribución de los números aleatorios generados en cada una de las columnas del DataFrame.
```
