```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('_mpl-gallery-nogrid')

# make data
x = df['NUM_VOTOS'].values
print (x)
partidos = df.index.values
colors = plt.get_cmap('Blues')(np.linspace(1.2, 0.4, len(x)))

# plot
fig, ax = plt.subplots()

a, b = ax.pie(x, colors=colors)
 
ax.legend(a, partidos,
          title="Los que nos roban",
          loc="center left",
          bbox_to_anchor=(2, 0, 0.5, 1))
plt.show()
```
