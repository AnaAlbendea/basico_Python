```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = np.random.choice(['sunny', 'rain', 'cloudy'], 1000, p=[1 / 4.0, 1 / 4.0, 1 / 2.0])
df = pd.DataFrame(data=data, columns=['weather'])

df['weather'].value_counts().plot(kind='bar')
plt.show()

#Este código genera aleatoriamente una lista de 1000 elementos 
#que representan el clima (soleado, lluvioso o nublado) con probabilidades de ocurrencia diferentes.
#Luego, crea un DataFrame con esta lista de datos y cuenta la frecuencia de cada tipo de clima. 
#Finalmente, muestra un gráfico de barras que representa la frecuencia de cada tipo de clima.
```
