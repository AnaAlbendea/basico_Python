```python
import matplotlib.pyplot as plt
import numpy as np
def f(t):#se crea una funcion que devuelve otra funcion
    return np.exp(-t) * np.cos(2*np.pi*t) 

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure() #? donde utiliza esto?
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
#este código muestra dos gráficos en una misma figura: el primero muestra los puntos de una función exponencial 
#y el segundo muestra la función coseno.
```
